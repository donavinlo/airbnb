```python
def get_host_time_length(date_compiled, form):
    "create column that counts the number of days a host has been with Airbnb"
    date_compiled = dt.datetime.strptime(date_compiled, '%Y-%m-%d')
    df['host_since'] = pd.to_datetime(df['host_since'], format='%Y-%m-%d')
    df['host_length'] = (date_compiled - df['host_since']).dt.days
    
```

```python
def clean_rates():
    "remove parentheses from strings"
    df['host_response_rate'] = df['host_response_rate'].str.replace("%", "").astype(float)
    df['host_acceptance_rate'] = df['host_acceptance_rate'].str.replace("%", "").astype(float)
```

```python
def clean_verification_list(val):
    "function for lambda to change from list in a string to list of strings"
    val = val.strip("[]").replace("'", "").replace(" ", "").strip(" ").split(",")
    if '' in val:
        val.remove('')
    if 'None' in val:
        val.remove('None')
    return val

def clean_verification():
    "Get count of verification"
    df['host_verifications'] = df['host_verifications'].apply(clean_verification_list)
    df['verification_count'] = df['host_verifications'].str.len()
```

```python
def bathroom_clean():
    "Create column for if bathroom shared or not. Also, separate count of bathroom"
    df['shared_rr'] = np.where(df['bathrooms_text'].str.contains(r'[sS]hared'), True, False)
    df['rr_count'] = df['bathrooms_text'].str.extract(r'(\d*\.?\d+)').astype(float)
```

```python
get_host_time_length('2022-03-20', form='%Y-%m-%d')
clean_rates()
clean_verification()
bathroom_clean()
```
